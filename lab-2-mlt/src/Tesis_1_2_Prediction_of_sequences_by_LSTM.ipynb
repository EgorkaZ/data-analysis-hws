## **PREDICTION OF SEQUENCE OF DAYS USING ONLY ONE VARIABLE**

In this algorithm data are collected from Binance Exchange and the prediction is about one sequence of days. The algorithm can be used for any active. In this case the active choosen has been Bitcoin. The feature is price close.

## Load data

```python
import pandas as pd
import ssl # we need to import this library and tweak one setting due to fact we use HTTPS certificate(s)
import yaml
```

```python
cfg = open("../config.yaml")
config = yaml.load(cfg, Loader=yaml.FullLoader)
```

```python
filepath = config["dataset_path"]
ssl._create_default_https_context = ssl._create_unverified_context

# Now we want to create a dataframe and use Pandas' to_csv function to read in our file
df = pd.read_csv(filepath, skiprows=1)  # we use skiprows parameter because first row contains our web address
```

```python
df = df.sort_values(['date'], ascending=[True])
df.head(15)  # print first 15 lines of dataframe
```

```python
column_to_predict = config["column_to_predict"]
epochs = config["epochs"]
batch_size = config["batch_size"]
test_ratio = config["test_ratio"]

look_back = config["look_back"]
look_forward = config["look_forward_2"]
num_back_X_plot = 4
```

## Data preparation

```python
import modules.data_prepare as prepare
```

```python
dataset, scaler = prepare.extract_and_scale_data(df, column_to_predict)
X, y = prepare.split_and_reshape(dataset, scaler, look_back, look_forward, test_ratio)
```

## Creating LSTM model

```python
from modules.helpers import create_model
model = create_model(X.train, y.train, look_back, epochs, batch_size)
```

## Executing the model

```python
from modules.helpers import predict_and_measure
```

### on train data

```python
train_predict, train_rmse = predict_and_measure(model, X.train, y.train, scaler)
print('Train Score: %.2f RMSE' % (train_rmse))
```

### on test data

```python
test_predict, test_rmse = predict_and_measure(model, X.test, y.test, scaler)
print('Test Score: %.2f RMSE' % (test_rmse))
```

## Plot the results

```python
prepare_plot(
    title="Historical Bitcoin Price (Binance Exchange)",
    xlabel="day",
    ylabel="Price"
)
plt.plot(dataset)
plt.show()
```

### Predictions

```python
import numpy as np
```

```python
first_test = len(X.train.den())
total_len = look_back + look_forward
```

```python
for i in range(0, 10):
    left_border = first_test + i

    # look ahead that was considered by the model
    real_plot = np.full((total_len), np.nan)
    real_plot[:look_back] = X.test.den()[i]

    # what the model should ideally predict
    real_preview = np.full((total_len), np.nan)
    real_preview[look_back:total_len] = dataset[left_border+look_back:left_border+total_len, 0]

    # what the model actually predicts
    predict_preview = np.full_like(real_plot, np.nan)
    predict_preview[-look_forward:] = test_predict[i]

    # connect previews to the solid plot
    real_preview[-look_forward-1] = real_plot[-look_forward-1]
    predict_preview[-look_forward-1] = real_plot[-look_forward-1]

    _, ax = plt.subplots()
    ax.plot(real_plot, '-', label='Historic X Price', color = 'orange')
    ax.plot(predict_preview, '--', label='Y Actual Predict', color = 'blue')
    ax.plot(real_preview, '--', label='Y Expected Predict', color = 'green')
    legend = ax.legend(loc='best', shadow=True, fontsize='medium')
    prepare_plot(
        title="Bitcoin Price daily time-frame three-days Predictions",
        xlabel="day",
        ylabel="Price"
    )
    plt.show()
```
