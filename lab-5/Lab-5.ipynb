1. Примените алгоритм кластеризации к спроецированным данным. Расчитайте средние внутрикластерное и межкласторное расстояния получившейся проекции
1. Опишите результаты 

```python
from math import sqrt

def dist_py(x1, x2, y1, y2):
    x = x2 - x1
    y = y2 - y1
    return sqrt(x * x + y * y)

def count_metrics_py(xs, ys, clusters):
    in_same_sum = 0.
    in_same_cnt = 0
    in_diff_sum = 0.
    in_diff_cnt = 0
    for (i, x1, y1, cl1) in zip(range(len(xs)), xs, ys, clusters):
        for (j, x2, y2, cl2) in zip(range(len(xs)), xs, ys, clusters):
            if j >= i:
                break
            if cl1 == cl2:
                in_same_sum += dist_py(x1, x2, y1, y2)
                in_same_cnt += 1
            else:
                in_diff_sum += dist_py(x1, x2, y1, y2)
                in_diff_cnt += 1
    return (in_same_sum / in_same_cnt, in_diff_sum / in_diff_cnt)
```
